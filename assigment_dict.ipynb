```python
Create a dictionary with 3 key-value pairs and print it.
```

```python
a={'banana':30,'carrot':40,'apple':50}
print(a)
```

```python
Access and print the value of a specific key from a dictionary.
```

```python
a={'banana':30,'carrot':40,'apple':50}
print(a['banana'])
print(a['apple'])
```

```python
Use get() to retrieve a value safely without causing an error.
```

```python
a={'banana':30,'carrot':40,'apple':50}
print(a.get('banana'))
```

```python
Merge two dictionaries into a new one using update().
```

```python
a={'banana':30,'carrot':40,'apple':50}
b={'name':'hari','age':40}
c=a.copy()
c.update(b)
print(c)
```

```python
Remove a specific key using pop() and print the updated dictionary.
```

```python
a={'banana':30,'carrot':40,'apple':50}
print(a.pop('carrot'))
print(a)
```

```python
Retrieve all the keys from a dictionary using keys() method.
```

```python
a={'banana':30,'carrot':40,'apple':50}
print(a.keys())
```

```python
 Retrieve all the values from a dictionary using values() method.
```

```python
a={'name':'hari','age':24,'salary':2999}
print(a.values())
```

```python
Get all key-value pairs from a dictionary using items().
```

```python
a={'name':'hari','age':24,'salary':2999}
print(a.items())
```

Use setdefault() to insert a key only if it doesnâ€™t exist.

```python
a={'name':'hari','age':25}
a.setdefault('salary',1200)

print(a)
```

Convert a list of tuples [("a",1),("b",2)] into a dictionary.


```python
a=[('a',1),('b',2)]
b=dict(a)
print(b)
```

Convert two lists ["a","b"] and [1,2] into a dictionary.


```python
a=["a","b"]
b=[1,2]
c=dict(zip(a,b))
print(c)
```

 Find the length of a dictionary using len().

```python
a={'name':'hari','age':25}
a.clear()
print(a)
```
