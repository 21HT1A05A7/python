```python
def sharuk(n):
  s = 0
  for j in range(1,n):
    if n % j == 0:
      s = s + j

  if s == n:
    print(f"{n} : perfect")
  else:
    print(f"{n} : Not Perfect")

sharuk(6) # calling
```
